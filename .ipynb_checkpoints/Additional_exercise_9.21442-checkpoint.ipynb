```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
## Excercise 1
. Create a figure object called fig using plt.figure()
. use add_axes to add an axis to the figure canvas at[0,0,1,1]. call this new axis ax.
. Plot (x,y) on that axes and set the labels and title to match the plot
```

```python
figure = plt.figure()
ax = figure.add_axes([0,0,1,1])
ax.plot(x, y, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
```
